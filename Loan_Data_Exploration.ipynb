```python
df_after09 = df[df.ListingCreationDate >'2009-07-01 11:46:18.690000000']
df_after09.info()
```

>According to the ProperLoans data dictionary there was change in the data entry where data is "Applicable for loans originated after July 2009." For example CreditGrade is pre 2009 and ProsperRating is post 2009. Removing the pre Jul 2009 data leave 84853 rows which is plenty representitive and will aid in data consistency.

```python
#Create smaller dataframe (df_total) of all the interesting features
df_total = df_after09[['ListingKey','Term','ProsperRating (Alpha)','ProsperScore','LoanOriginalAmount','LenderYield','AmountDelinquent','LP_CustomerPayments','LP_InterestandFees','EstimatedReturn','EstimatedLoss','EstimatedEffectiveYield','LoanStatus','BorrowerAPR','BorrowerRate','BorrowerState','Occupation','IncomeRange','StatedMonthlyIncome','MonthlyLoanPayment']]
df_total.head()
```

```python
#Shorten the name of 'ProsperRating (Alpha)'
df_total.rename(columns={'ProsperRating (Alpha)': 'ProsperRating'}, inplace=True)
```

```python
#Interesting quantitative values
df_quant = df_total[['ListingKey','EstimatedReturn','EstimatedLoss','EstimatedEffectiveYield','StatedMonthlyIncome','MonthlyLoanPayment','LoanOriginalAmount','LP_CustomerPayments']]
df_quant.sample(10)
```

```python
print(df_quant.EstimatedReturn.mean())
print(df_quant.EstimatedLoss.mean())
print(df_quant.EstimatedEffectiveYield.mean())
print(df_quant.LoanOriginalAmount.mean())
```

```python
df_quant.hist(figsize=[15,15],bins=15);
```

>A quick histogram on the quantitative values of interest reveals a few unexpected insights. The EstimatedLoss and EstimatedReturn seems typical. Estimated return has a normal distribution with a mean just below 10% and EstimatedLoss a right skewed distribution with a mean of about 8%. The LoanOriginalAmount has large spread in values but a typical loan in this dataset is 8K.

```python
df_quant.StatedMonthlyIncome.describe()
```

```python
#Create a column call YieldAmount which is the loanAmount times the LenderYield
df_total['YieldAmount'] = df_total.LenderYield * df_total.LoanOriginalAmount
df_total['EstimatedYieldAmount'] = df_total.EstimatedEffectiveYield * df_total.LoanOriginalAmount
df_total.YieldAmount.describe()
df_total[['EstimatedYieldAmount','YieldAmount']].describe()
```
